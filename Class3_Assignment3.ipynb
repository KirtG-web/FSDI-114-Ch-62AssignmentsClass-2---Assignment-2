Mini-Challenge #5 -- peek() and size() methods

1. Now add two more methods:
    - peek() --> return the top item
    - size() --> return the number of items
    

```python
# Step 1: Define the stack class
class Stack:
    def __init__(self):
        self.items = []  # empty list represents the stack 

    def is_empty(self):
        return len(self.items) == 0

# Define the pop() method
    def pop(self):
        if self.is_empty():
            return None
        return self.items.pop()   

# Step 1: peek() method
    def peek(self):
        if self.is_empty():
            return None
        return self.items[-1]

# Step 2: size() method
    def size(self):
        return len(self.items)

# Step 2: Define push() method
    def push(self, item):
        self.items.append(item)  # add item to the top of the stack

s = Stack()

s.push(5)
s.push(10)
s.push(15)

print("Stack peek: ", s.peek())
print("Stack pop: ", s.pop())
print("Stack size: ", s.size())

```

# Testing the Queue Class


```python
class Queue:
    def __init__(self):
        self.items = []  # empty list represents the queue

    def is_empty(self):
        return len(self.items) == 0

    # Add item to the back of the queue
    def enqueue(self, item):
        self.items.append(item)

    # Remove item from the front of the queue
    def dequeue(self):
        if self.is_empty():
            return None
        return self.items.pop(0)
    
    # Return the size of the queue
    def size(self):
        return len(self.items)

# Testing the Queue Class
q = Queue()

q.enqueue("A")
q.enqueue("B")
q.enqueue("C")

print("Queue dequeue:", q.dequeue())  # Should print "A"
print("Queue size:", q.size())        # Should print 2
```
