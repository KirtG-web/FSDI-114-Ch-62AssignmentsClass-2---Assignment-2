Given a list of numbers, determine which one is the largest.
    1. For example given numbs = [12,4,27, 3, 188, 50, 7], the output shoud be 188.
    2. Write the pseudocode.
    3. Submit the pseduocode in a markdown cell in your Jupyter Notebook

    4. SET largest TO first element in the list

FOR each number in the list
    IF number > largest
        SET largest TO number
    END IF
END FOR

OUTPUT largest

[12, 4, 27, 3, 188, 50, 7]



```python
nums = [12, 4, 27, 1000, 50, 7]

def find_largest(nums):
    if not nums:
        return "No values to compare"

    largest = nums[0]

    for n in nums[1:]:
        if n > largest:
            largest = n

    return largest   # â† must be indented inside the function

print(find_largest(nums))
```

```python
### Pseudocode: Find the Largest Number

numbers = [12, 4, 27, 3, 188, 50, 7]

largest = numbers[0]

for num in numbers:
    if num > largest:
        largest = num

print(largest)

```
