```python
# Challengs 4. Reverse a word using a stack
# Use a stack to invert the word "Hello"
stack=[]
word = "Hello"

# Push each character onto the stack

for char in word:
        stack.append(char)

# Pop to reverse the word

reversed_word = ""
while stack:
    reversed_word += stack.pop()

print(reversed_word)
```

```python
# Queue looking at the front or head of the queue and looking at the back/tail/rear of the queue.  At the front is Dequeue and at the end is Enqueue.

# Challenge 5 Enqueue three items-Solution  Instructions: 1. Create an empty queue. 2. Enqueue the values 10, 20, and 30 in that order. 3. Then print the queue.
from collections import deque

# 1. Create an empty queue
queue = deque()

# 2. Enqueue the values 10, 20, and 30 in that order
queue.append(10)
queue.append(20)
queue.append(30)

# 3. Print the queue

print("Queue:", list(queue))
# Output: Queue: [10, 20, 30]
```
